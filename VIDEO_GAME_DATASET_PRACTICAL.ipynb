```python
import matplotlib.pyplot as plt

# Group by 'Publisher' and calculate the sum of 'total_sales'
grouped = df.groupby('publisher')['total_sales'].sum()

# Create a bar plot
plt.bar(grouped.index, grouped.values)
plt.xlabel('Publisher')
plt.ylabel('Total Sales')
plt.title('Total Sales by Publisher')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'Publisher' and calculate the sum of 'total_sales'
grouped = df.groupby('publisher')['total_sales'].sum().sort_values(ascending=False)[:10]

# Create a bar plot
plt.bar(grouped.index, grouped.values)
plt.xlabel('Publisher')
plt.ylabel('Total Sales')
plt.title('Total Sales by Publisher (Top 10)')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by 'console' and calculate the sum of 'total_sales'
grouped = df.groupby('console')['total_sales'].sum().sort_values(ascending=False)[:10]

# Create a scatter plot
plt.scatter (grouped.index, grouped.values)
plt.xlabel('console')
plt.ylabel('Total Sales')
plt.title('Total Sales by console (Top 10)')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
